```python
import pandas as pd
```

```python
pd.options.mode.chained_assignment = None
```

```python
df = pd.read_csv("data/aoa_assoc_word2vec_colex.csv")
```

```python
df_colex = pd.read_csv("data/df_all_raw.csv")
```

```python
df_colex.columns = list(map(str.lower, df_colex.columns))
df_colex = df_colex.drop(columns=['dataset_id', 'form_id', 'form', 'gloss_in_source', 'iso639p3code', 'mrc_word', 'kucera_francis_frequency'])
```

```python
df_colex.head()
```

```python
df_selected_columns = df_colex[["concreteness", "concepticon_gloss", "concepticon_id"]]
```

```python
df.head()
```

```python
df_selected_columns.head()
```

```python
concept1 = df["concept_1"]
concept2 = df["concept_2"]
```

```python
my_dict = dict(zip(df_selected_columns["concepticon_gloss"], df_selected_columns["concreteness"]))
my_dict["GOLD"]
```

```python
concrete_diff = []
concrete_1 = []
concrete_2 = []
for c1, c2 in zip(concept1, concept2):
    concrete1 = my_dict[c1]
    concrete2 = my_dict[c2]
    val = abs(concrete1 - concrete2)
    concrete_1.append(concrete1)
    concrete_2.append(concrete2)
    concrete_diff.append(val)
```

```python
df['concreteness_diff'] = concrete_diff
df['concreteness_1'] = concrete_1
df['concreteness_2'] = concrete_2
```

```python
print("Number of rows is NaN:")
print(len(df[df['concreteness_diff'].isnull()]))
print("Number of rows is not NaN:")
print(len(df[df['concreteness_diff'].notnull()]))
```

```python
df = df[df['concreteness_diff'].notnull()]
df.head()
```

```python
df.to_csv('data/final_dataset.csv', index=False)
```
